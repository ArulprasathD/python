## PYTHON LOGIC OPERATORS :

## and , or , not

```python
a = 10
b=5
print(a>b and a<b )
```

```python
print(a>b or a<b)
```

```python
j =12

print(j>8 and j<16)
```

```python
print(not(j<3 or j>4))
```

#  Membership Operators

## in , not in

```python
a = "besent"

b = "besent"

print(a in b)
```

```python
print( "s" in a)
```

```python
print("besent" in b)
```

```python
print("besent" not in b)
```

```python
print(not("technology" in a))
```
