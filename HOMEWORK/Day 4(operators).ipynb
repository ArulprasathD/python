```python
# Arithmetic operators

num_1 = int(input("Enter the value 1 : "))
num_2 = int(input("Enter the value 2 : "))
        
                  
```

```python
print(num_1 + num_2)    #add
```

```python
print(num_1 - num_2)    #sub
```

```python
print(num_1 * num_2)    #multiple
```

```python
print(num_1 % num_2)    #modules
```

```python
print(num_1 / num_2)    #float
```

```python
print(num_1 // num_2)    #double float
```
