```python
l = [1,2,3,4,3,2,1,2,5,5,5,5]

duplicates = []


g = l.count(1)
w = l.count(2)
e = l.count(3)
r = l.count(4)
t = l.count(5)
print("1 :",g)
print("2 :",w)
print("3 :",e)
print("4 :",r)
print("5 :",t)

print("----duplicates using for loop----")
for i in l:
    if l.count(i)>1 and i not in duplicates:
        print(f"{i} : {l.count(i)}")
        duplicates.append(i)
    
```
