## TYPE CASTING

### Str to other data types :

```python
a = "12345"

print(type(a))

print(int(a))   #STR to INT

print(float(a)) #STR to FLOAT

print(complex(a)) #STR to COMPLEX

print(bool(a))  #STR to BOOLEAN
```

## int to other data types :

```python
b = 12
u = 33
print(type(b))

print(str(b) + str(u)) #int to str

print(bool(b))  #int to boolean 

print(complex(b)) #int to complex

print(float(b))  #int to float
```

## float to other data types :

```python
i = 33.33

print(type(i))

print(int(i)) #float to int

print(str(i)) #float to str

print(complex(i))  #float to complex

print(bool(i)) #float to bool
```

## boolean to other data types

```python
a = True
b = False

print(f"the data type of a,b {type(a)} and {type(b)}")


print(int(a))
print(int(b))
print("------------------")
print(str(a))
print(str(b))
print("------------------")
print(complex(a))
print(complex(b))
print("------------------")
print(float(a))
print(float(b))
```

##  complex to other data types

```python
a = 3 + 7j
# = real + imaginary 
# = a +  bj

print(type(a))

print(str(a))
print("--------------------------")
print(f"real part : {int(a.real)}")   #real (attribute)  
print(f"imaginary part : {int(a.imag)}")  #imaginary (attribute)
print("--------------------------")

print(f"real part : {float(a.real)}")
print(f"imaginary part : {float(a.imag)}")  #imaginary
print("--------------------------")
print(bool(a))
```
