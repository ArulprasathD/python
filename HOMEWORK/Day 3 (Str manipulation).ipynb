## STRING MANIPULATION METHODS

```python
j = "Yogesh waran"

print(j.upper())                                 #uppercase

print(j.lower())                                 #lowercase

print(j.title())                                 #capitalize the first letter in each word

print(j.split("g"),j.split("r"))                 #split str to list

print(j.capitalize())                            #first letter capitalize

update_j = j.replace("Yogesh waran","vinnu")     #replace once to another
print(update_j)

print(update_j.find ("n"))                       #finding index value

#print(update_j.split())
```

##  _________________________________________________________________________________________________________
REMOVE SPACE

```python
y = "           arul             "

alter = y.rstrip()
print(f"my name is {alter},i'm from aaaa")    #remove space in right side
alter = y.lstrip()
print(f"my name is {alter},i'm from aaaa")    #remove space in left side
alter = y.strip()
print(f"my name is {alter},i'm from aaaa")    #remove space in both side
```
