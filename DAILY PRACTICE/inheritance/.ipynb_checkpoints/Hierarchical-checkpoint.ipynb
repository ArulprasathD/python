```python
class besent:
    def course_center(self):
        print("Besent technology")

class staff1(besent):
    def data_science(self):
        print("DATA SCIENCE class at 8.00 am")

class staff2(besent):
    def full_stack(self):
        print("FULL STACK class at 9.00 am")

j = staff1()
j.course_center()
j.data_science()

print("------------------------------")

k = staff2()
k.course_center()
k.full_stack()
```
