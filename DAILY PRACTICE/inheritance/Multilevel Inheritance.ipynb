```python
class grandfather:      #parent
    def house(self):
        print("grandpah : I have a 3BHK house")

class father(grandfather):        #child
    def shop(self):
        print("father : I have a Rice mill")

class son(father):                #child class
    def car(self):
        print("son : I have a lambo")


s1 = son()
s1.car()
s1.house()
s1.shop()
```

```python
class animal:
    def eat(self):
        print("Animal is eating")

class dog(animal):
    def bark(self):
        print("Dog is barking")

class babydog(dog):
    def weep(self):
        print("Baby dog is weeping")

a1 = animal()
```
