```python
#check prime number

val = int(input("Enter number : "))

if val <= 1:
    print("Not Prime")

else:
    for i in range(2,val):  #int(val**0.5)+1
        if val % i == 0:
            print("Not prime")
            break
    else:
            print("Prime")
```

```python
#Factorial program

num = int(input("Enter Number : "))

fact = 1

for i in range(1,num+1):
    fact *= i

print(fact)
```
