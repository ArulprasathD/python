```python
"""Print numbers from 1 to 20.
For multiples of 3, print "Fizz" instead of the number."""


for i in range(1,21):
    if i%3 ==0:
        print("Fizz")
    else:
        print(i)
    
```

```python
"""Next small upgrade question:

ðŸ‘‰ What if I ask:
Multiples of 3 â†’ Fizz
Multiples of 5 â†’ Buzz
Multiples of both â†’ FizzBuzz"""


for i in range(1,21):
    if i % 3==0 and i % 5==0:
        print("FizzBuzz")
    elif i % 5 == 0:
        print("Buzz")
    elif i % 3==0 :
        print("Fizz")
    else:
        print(i)
```

```python
"""ðŸ‘‰ Print from 1 to 50
But:

Multiple of 3 â†’ Fizz

Multiple of 5 â†’ Buzz

Multiple of 7 â†’ Boom

Multiple of all three â†’ FizzBuzzBoom"""


for i in range(1,50+1):
    if i % 3==0 and i % 5==0 and i % 7 ==0:
        print("FizzBuzzBoom")
    elif i % 7 == 0:
        print("Boom")
    elif i % 5 == 0:
        print("Buzz")
    elif i % 3==0 :
        print("Fizz")
    else:
        print(i)
        
```

```python
"""Find the largest number in a list without using max()"""

nums = [10, 45, 3, 99, 23, 67]

largest = 0

for i in nums:
    if largest <= i:
        largest =  i
print(largest)
```
