```python
i = 9875

total = 0

while i > 0:
    digit = i%10
    total += digit
    i = i//10
    
print("sum of digit :",total)
```

```python
i = 1234

reverse = 0

while i > 0:
    digit = i % 10
    reverse = reverse*10 + digit
    i = i // 10
print(reverse)
```
