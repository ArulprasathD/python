```python
#2 Count vowels in a string

a = "orange"
count = 0

for b in  a.lower():
    print(b,end = "")
    if b in "aeiou":
        count += 1

#print(f""" 
#Count is : {count}
#      """)

print("\ncount is ",count)
```

```python
#3

a = "level"

if a.lower() == a.lower()[::-1]:
    print("Palindrome")
else:
    print("not Palindrome")
```

```python
#4

a = "   apple     "
print("before :",a)
print(len(a))
b = a.strip()
print(b)
print(len(b))
c = a.rstrip()
print(c)
print(len(c))
d = a.lstrip()
print(d)
print(len(d))
```
