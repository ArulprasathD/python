```python
#7 ATM withdraw logic

atm_pin = "3275"
attempt = 0

while attempt < 3:
    enter_pin = input("Enter your PIN: ")

    if enter_pin == atm_pin:
        print("Authentication successful")
        break
    else:
        attempt += 1
        print("Wrong PIN")

if attempt == 3:
    print("Your account is blocked")
```

```python
#12 Simple calculator using if-elif.

num1 = int(input("number 1 : "))
num2 = int(input("number 2 : "))
operator = input("{add | sub | mul | div or + | - | * | / } : ") .lower()

if "add" == operator or "+" == operator:
    print("add",num1+num2)

elif "sub" == operator or "-" == operator:
    print("sub",num1-num2)

elif "mul" == operator or "*" == operator:
    print("mul",num1*num2)
    
elif "div" == operator or "/" == operator:
    if num2 != 0:
        print("div",num1/num2)
    else:
        print("cannot divide by zero")
else:
    print("invalid operator")

        
```

```python
#14 Temperature condition example


temp = int(input("enter the current temperature level : "))


if temp > 40:
    print(f"{temp}°C current climate very hot")
elif 40 >= temp and  31 <= temp:
    print(f"{temp}°C current climate hot")
elif 30 >= temp and 21 <= temp:
    print(f"{temp}°C current climate normal")
elif 20 >= temp and 11 <= temp:
    print(f"{temp}°C current climate cold")
else:
    print(f"{temp}°C current climate very cold")
```
