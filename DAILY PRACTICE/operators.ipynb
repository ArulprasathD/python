```python
#1

num1 = 15
num2 = 22

add = num1+num2
print("add   :",add)

sub = num1-num2
print("sub   :",sub)

mul = num1*num2
print("mul   :",mul)

div = num1/num2
print("div   :",div)

d_fl = num1//num2
print("floor :",d_fl)

mod = num1%num2
print("mod   :",mod)
```

```python
#3 power opwerator

val_1 = int(input("enter the value : "))
val_2 = int(input("enter the value : ")) #anything power 0 = 1


c = val_1 ** val_2
print(c)
```

```python
#4 Comparison operators example (==, !=, >, <, >=, <=)

a = int(input("enter the number 1 : ".title()))
b = int(input("enter the number 2 : ".title()))

print( a == b )

print(a != b)

print(a > b)

print(a < b)

print(a >= b)

print(a <= b)
```

```python
#5 Logical operators (and, or, not)

a = 2
b = 4

print(a>b and b<a)

print(a>b or a<b)

print(not(a>b))
```

```python
#6 Assignment operators (+=, -=, *=, /=)


a = 5

a += 2
print(a)

a -= 3
print(a)


a *= 3
print(a)

a /= 2
print(a)
```

```python
#7 Identity operators (is, is not)

a = 257
b = 257
print(a is b)

a = 256
b = 256
print(a is b)

a = 10
b = 20
print(not(a is b))
```

```python
#8 Membership operators (in, not in)

book = "Harry potter and the prison of stone"

print("prison" in book)
print(not("prison" in  book))
```

```python
#9 and 10 >> Bitwise AND or example

a = 12
b = 13

print(a & b)
print(a | b)
```
