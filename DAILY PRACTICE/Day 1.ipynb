# IF CONDITIONS

```python
# Write a program to check whether a number is positive

num = int(input("Enter the number :"))         


if num > 0:
    print(f"{num} is positive")
elif num < 0 :
    print(f"{num} is negative")
else:
    print(f"{num} is zero")
```

```python
#Write a program to check whether a number is even or odd.

a = int(input("enter the number : "))

if a % 2 == 0:
    print(f"{a} is even ")
else:
    print(f"{a} is odd ")
```

```python
#Write a program to assign grades based on marks:

mark = int(input("Enter your mark: "))

if mark > 0 or mark < 100:
    print("Invalid mark")

elif mark >= 90:
    print("Pass")
    print("Grade 'A'")

elif mark >= 75:
    print("Pass")
    print("Grade 'B'")

elif mark >= 50:
    print("Pass")
    print("Grade 'C'")

else:
    print("Fail")
```

```python
#Create a simple calculator program

num_1 = float(input("Enter the number 1 : "))
num_2 = float(input("Enter the number 2 : "))
operator =input("operator : " )  #that should be (add,sub,mul,div (or) +,-,*,/)

if operator == "add" or operator == "+":
    print(f"{num_1} + {num_2} = {num_1+num_2}")
elif operator == "sub" or operator == "-":
    print(f"{num_1} - {num_2} = {num_1-num_2}")
elif operator == "mul" or operator == "*":
    print(f"{num_1} * {num_2} = {num_1*num_2}")
elif operator == "div" or operator == "/":
     if num_2 != 0:  
        print(f"{num_1} / {num_2} = {num_1 / num_2}")
     else:
        print("Cannot divide by zero")
else:
    print("invalid operator")

```
