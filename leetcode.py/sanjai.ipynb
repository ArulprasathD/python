```python
#remove duplicate from list

a = [1,2,2,3,3,1,4,5]

after_remove = []

for i in a:
    if i not in after_remove:
        after_remove.append(i)

print(after_remove)
```

```python
#odd r even

a = int(input("Enter number : "))

if a % 2 == 0:
    print("Even")

else:
    print("Odd")
```
